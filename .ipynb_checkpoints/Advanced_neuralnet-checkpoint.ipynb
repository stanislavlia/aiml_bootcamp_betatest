## EX01

```python
def display_img(n, image_labels, image_pixels_tensor, y_predicts=None):
    i = 0
    
    num_images = len(image_pixels_tensor)
    end_index = min(i + n, num_images)
    num_rows = (end_index - i + 3) // 4  # Calculate number of rows required

    fig, axes = plt.subplots(num_rows, 4, figsize=(12, 3*num_rows))

    for j in range(i, end_index):
        row = (j - i) // 4
        col = (j - i) % 4
        axes[row, col].imshow(image_pixels_tensor[j] / 255, cmap="gray")
        axes[row, col].set_title(f"Label for image: {int(image_labels[j])}", fontsize=14)
        if (y_predicts is not None):
            axes[row, col].set_title(f"Label for image: {int(image_labels[j])}\nPrediction for image: {int(y_predicts[j])}", fontsize=14)
            
        axes[row, col].axis('off')

    # Hide any empty subplots
    for j in range(end_index - i, num_rows*4):
        row = j // 4
        col = j % 4
        axes[row, col].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
#Observing 
display_img(16, image_labels, image_pixels_tensor)
```

## EX02

#### Designing Neural Net architecture

### Using Logistic Regression for Binary Classification

```python
class LogisticRegression():
    def __init__(self, lr_rate=0.01, l2_reg=0):
        self.lr_rate = lr_rate
        self.l2_reg = l2_reg

        self.W = None
        self.b = None

        self.Z = None
        self.A = None

        self.dW = None
        self.db = None

        self.epsilon = 10 ** -6

        self.loss_history = []
    
    def _sigmoid(self, Z):
        return 1 / (1 + np.exp(-Z + self.epsilon))
    
    def _init_params(self, X):
        self.W = np.random.randn(X.shape[1], 1)
        self.b = np.random.random()
    
    def _compute_loss(self, y, y_pred):
        
        costs = -y * np.log(y_pred +  self.epsilon) - (1 - y)*np.log(1 - y_pred + self.epsilon)
        loss = np.mean(costs)
        return loss

    def _forward(self, X):
        self.Z = (X @ self.W) + self.b
        self.A = self._sigmoid(self.Z)

        return self.A

    def _backward(self, y, y_pred, X):

        y = y.reshape(-1, 1)
        y_pred = y_pred.reshape(-1, 1)
   
        loss = self._compute_loss(y, y_pred)

        #computing gradients
            
        self.d_W = X.T @ (y_pred - y)
        self.d_b = np.mean(y_pred - y)
        

        #updating weights
        self.W = self.W - self.lr_rate * self.d_W
        self.b = self.b - self.lr_rate * self.d_b


    def fit(self, X_train, y_train, epochs=10, batch_size=100):

        self._init_params(X_train)

        for epoch in range(epochs):

        
            
            for batch_num in range(math.ceil(len(X_train) / batch_size)):

                X_batch = X_train[batch_num * batch_size : (1 + batch_num) * batch_size]
                y_batch = y_train[batch_num * batch_size : (1 + batch_num) * batch_size].reshape(-1, 1)
                
                

                #compute predictions
                y_pred = self._forward(X_batch)

                #compute loss
                batch_loss = self._compute_loss(y_batch, y_pred)

                self._backward(y_batch, y_pred, X_batch)
            print(f"EPOCH {epoch}; Loss = {batch_loss}")

            self.loss_history.append(batch_loss)            
    def predict(self, X):

        y_preds = self._forward(X)

        return np.round(y_preds)
        
```

### Train/Test split



```python
X_train, y_train = image_pixels_tensor_flatten[:7000], image_labels[:7000]
X_test, y_test = image_pixels_tensor_flatten[7000:], image_labels[7000:]

#Normalizing the data
X_train, X_test = X_train / 255.0 , X_test / 255.0
```

### Training model



```python
logreg = LogisticRegression(lr_rate=0.0003)

logreg.fit(X_train, y_train, epochs=50, batch_size=256)
```

### Evaluation & Analysis

### EX03

```python
def evaluate_model(model, X, y):

    y_predicts = model.predict(X)
    y = y.reshape(-1, 1)
   
    accuracy = np.sum((y_predicts == y)) / len(y)

    return (accuracy)


def plot_history(history):
    plt.figure(figsize=(14, 7))
    plt.title("Learning History")
    plt.plot(np.arange(0, len(history), 1) ,history, marker="o", label="Train Loss")
    plt.xlabel("Training epochs", fontsize=16)
    plt.ylabel("Loss function", fontsize=16)
        

```

```python
train_accuracy = round(evaluate_model(logreg, X_train, y_train), 4)
test_accuracy = round(evaluate_model(logreg, X_test, y_test), 4)

print(f"Train Accuracy of our model: {train_accuracy * 100}%")
print(f"Test Accuracy of our model: {test_accuracy * 100}%")

```

```python
plot_history(logreg.loss_history)
```

### Visualizing missclassified images


```python
def get_misclassified_indexes(y_true, y_pred):
    y_true = y_true.reshape(-1, )
    y_pred = y_pred.reshape(-1, )
    
    misclassified_indexes = []
    for i in range(len(y_true)):
        if y_true[i] != y_pred[i]:

            misclassified_indexes.append(i)
    return misclassified_indexes


y_test_predictions = logreg.predict(X_test)

misclassified_indexes = get_misclassified_indexes(y_test, y_test_predictions)

actual_labels = y_test[misclassified_indexes]
wrong_predicts = y_test_predictions[misclassified_indexes]
wrong_images = X_test[misclassified_indexes].reshape(-1, 28, 28) * 255 #reverse transformation


display_img(16, actual_labels, wrong_images, wrong_predicts)
```
